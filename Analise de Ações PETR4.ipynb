#Customizando Faixas de Eixos (Range)

```
import plotly.graph_objs as go

Close = go.Scatter(
                x=df.Date,
                y=df.Close,
                name = "PETR4 Close",
                line = dict(color = '#17BECF'))

layout = dict(
    title = "Faixa de Data Customizada",
    title_x=0.5,
    xaxis = dict(
        range = ['2017-01-01','2017-12-31']
    )
)

figura = dict(data=data, layout=layout)
py.iplot(figura, filename = "Manually Set Range")
```

#Serie com Rangeslider e Botoes


```
import plotly.graph_objs as go

Close = go.Scatter(
                x=df.Date,
                y=df.Close,
                name = "PETR4 Close",
                line = dict(color = '#17BECF'))

layout = dict(
    title = "Serie com Rangeslider e Botoes",
    title_x=0.5,
    xaxis = dict(
        rangeseletor=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'), 
                dict(step='all')                  
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'

    )
)
```

```
trace = go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])

layout = go.Layout(
    title = 'Preços em Candlesticks',
    title_x = 0.5,
    xaxis = dict(
        rangeslider = dict(
            visible = False
        ) 
    )
)

data = [trace]

fig = go.Figure(data=data, layout=layout)      
```

```
#Media simples de 9 dias
df['MM_9'] = df.Close.rolling(window=9).mean()

#Media simples de 17 dias
df['MM_17'] = df.Close.rolling(window=17).mean()

```

```
#Media simples de 3 dias
df['MM_3'] = df.Close.rolling(window=3).mean()

#Media simples de 9 dias
df['MM_9'] = df.Close.rolling(window=9).mean()

#Media simples de 17 dias
df['MM_17'] = df.Close.rolling(window=17).mean()
```

```
Close = go.Scatter(
                x=df.Date,
                y=df.Close,
                name = "PETR4 Close",
                line = dict(color = '#330000'))

MM_3 = go.Scatter(
                x=df.Date,
                y=df['MM_3'],
                name = "Média Móvel 3 Periodos",
                line = dict(color = '#b2ffaa'))

MM_9 = go.Scatter(
                x=df.Date,
                y=df['MM_9'],
                name = "Média Móvel 9 Periodos",
                line = dict(color = '#cc11ff'))

MM_17 = go.Scatter(
                x=df.Date,
                y=df['MM_17'],
                name = "Média Móvel 17 Periodos",
                line = dict(color = '#22ffff'))

data = [Close, MM_3, MM_9, MM_17]
py.iplot(data)
```
